### 1번 퀘스트

이름, 부서, 급여 정보를 가진 데이터프레임에서
groupby()를 사용하여 부서별 급여 합계를 구하는 코드를 작성하세요.


```python
import pandas as pd

data = {
    '이름': ['홍길동', '김철수', '박영희', '이순신'],
    '부서': ['영업', '영업', '인사', '인사'],
    '급여': [5000, 5500, 4800, 5100]
}

df = pd. DataFrame(data)
print(df.groupby('부서')['급여'].sum())
```

### 2번 퀘스트

groupby()와 agg()를 사용하여 부서별 급여의 합계(sum)와 평균(mean)을
계산하는 코드를 작성하세요.


```python
import pandas as pd

data = {
    '이름': ['홍길동', '김철수', '박영희', '이순신', '강감찬', '신사임당'],
    '부서': ['영업','영업', '인사', '인사', 'IT', 'IT'],
    '급여': [5000,5500, 4800, 5100, 6000, 6200]
    }

df = pd.DataFrame(data)
print(df.groupby('부서')['급여'].agg(['sum', 'mean']))
```

### 3번 퀘스트

부서별로 데이터를 그룹화한 뒤,
각 부서의 평균 급여가 5000 이상인 경우만 필터링하는 코드를 작성하세요.


```python
import pandas as pd

data = {
    '이름': ['홍길동', '김철수', '박영희', '이순신', '강감찬', '신사임당'],
    '부서': ['영업', '영업', '인사', '인사', 'IT', 'IT'],
    '급여': [5000, 5500, 4800, 5100, 6000, 6200]
}

df = pd.DataFrame(data)
print(df.groupby('부서').filter(lambda x: x['급여'].mean() >= 5000))
```
