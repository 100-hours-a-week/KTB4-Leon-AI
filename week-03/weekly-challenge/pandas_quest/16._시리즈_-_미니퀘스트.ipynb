### 1번 퀘스트

Pandas의 Series를 리스트 [5, 10, 15, 20]을 사용하여 생성하고,
Series의 인덱스를 확인하는 코드를 작성하세요.


```python
import pandas as pd

list = pd.Series([5, 10, 15, 20])

print(list.index)
```

### 2번 퀘스트

딕셔너리 {'a': 100, 'b': 200, 'c': 300}을 이용하여 Pandas Series를 생성하고,
인덱스를 활용하여 'b'의 값을 출력하는 코드를 작성하세요.


```python
import pandas as pd
d = {'a': 100, 'b': 200, 'c': 300}
s = pd.Series(d)
print(s['b'])
```

### 3번 퀘스트

Pandas Series [1, 2, None, 4, None, 6]에 대해 결측값(NaN)을 확인하고,
모든 결측값을 0으로 채운 후 Series의 값을 출력하는 코드를 작성하세요.


```python
import pandas as pd

series = pd.Series([1, 2, None, 4, None, 6])
print(series.isnull())
print(series.fillna(0).values)
```
