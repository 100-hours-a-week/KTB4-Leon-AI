## 결측치 처리 - 미니퀘스트


### 1번 퀘스트

결측치가 포함된 데이터프레임을 만들고,
결측치를 탐지한 뒤 열별 결측치 개수를 출력하는 코드를 작성하세요.


```python
import pandas as pd
import numpy as np

data = {'이름': ['홍길동', '김철수', np.nan, '이영희'], '나이': [25, np.nan, 30, 28], '성별': ['남', '남', '여', np.nan]}
df = pd.DataFrame(data)
print(df.isnull().sum())
```

### 2번 퀘스트

결측치가 포함된 행을 삭제하고,
새로운 데이터프레임을 출력하는 코드를 작성하세요.


```python
import pandas as pd
import numpy as np

data = {'이름': ['홍길동', '김철수', np.nan, '이영희'], '나이': [25, np.nan, 30, 28], '성별': ['남', '남', '여', np.nan]}
df = pd.DataFrame(data)
print(df.dropna())
```

### 3번 퀘스트

결측치가 포함된 '나이' 열을 평균값으로 대체하고,
새로운 데이터프레임을 출력하는 코드를 작성하세요.


```python
import pandas as pd
import numpy as np

data = {'이름': ['홍길동', '김철수', np.nan, '이영희'], '나이': [25, np.nan, 30, 28], '성별': ['남', '남', '여', np.nan]}
df = pd.DataFrame(data)
df['나이'] = df['나이'].fillna(df['나이'].mean())
print(df)
```
