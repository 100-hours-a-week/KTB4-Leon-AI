## 문자열 처리 - 미니퀘스트


### 1번 퀘스트

Pandas의 .str 접근자를 사용하여
Series ["HELLO", "WORLD", "PYTHON", "PANDAS"]의 모든 문자열을 소문자로 변환하세요.


```python
import pandas as pd
data = pd.Series(["HELLO", "WORLD", "PYTHON", "PANDAS"])
print(data.str.lower())
```

### 2번 퀘스트

데이터프레임의 '이름' 컬럼에서 문자열 앞뒤 공백을 제거하고,
특정 문자열 "Doe"가 포함된 행을 필터링하는 코드를 작성하세요.


```python
import pandas as pd
df = pd.DataFrame({"이름": [" John Doe ", "Alice ", " Bob", "Charlie Doe "]})
df['이름'] = df['이름'].str.strip()
print(df[df['이름'].str.contains('Doe')])
```

### 3번 퀘스트

데이터프레임의 '설명' 컬럼 문자열을 공백 기준으로 나누고,
각 단어의 첫 글자만 추출하여 새로운 컬럼 '약어'를 생성하는 코드를 작성하세요.


```python
import pandas as pd
df = pd.DataFrame({"설명": ["빅데이터 분석","데이터 과학", "머신 러닝", "딥 러닝"]})
df['약어'] = df['설명'].apply(lambda x: ''.join([w[0] for w in x.split()]))
```
