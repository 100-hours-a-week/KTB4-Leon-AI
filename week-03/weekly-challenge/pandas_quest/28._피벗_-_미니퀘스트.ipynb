## 피벗 - 미니퀘스트


### 1번 퀘스트

날짜, 제품, 판매량 정보를 가진 데이터프레임을 만들고,
pivot()을 사용하여 날짜를 행 인덱스로, 제품을 열로,
판매량을 값으로 설정하는 코드를 작성하세요.


```python
import pandas as pd

data = {'날짜': ['2024-01-01', '2024-01-01', '2024-01-02', '2024-01-02'], '제품': ['A', 'B', 'A', 'B'], '판매량': [100, 200, 150, 250]}
df = pd.DataFrame(data)
print(df.pivot(index='날짜', columns='제품', values='판매량'))
```

### 2번 퀘스트

pivot_table()을 사용하여 카테고리를 행 인덱스로, 제품을 열로 설정하고,
판매량의 합계를 출력하는 코드를 작성하세요.


```python
import pandas as pd

data = {'카테고리': ['전자', '가전', '전자', '가전'], '제품': ['A', 'B', 'A', 'B'], '판매량': [100, 200, 150, 250]}
df = pd.DataFrame(data)
print(df.pivot_table(index='카테고리', columns='제품', values = '판매량', aggfunc='sum'))
```

### 3번 퀘스트

pivot()을 사용하여 날짜를 행으로, 제품을 열로 설정하고,
판매량과 이익을 동시에 피벗하는 코드를 작성하세요.


```python
import pandas as pd

data = {'날짜': ['2024-01-01', '2024-01-01', '2024-01-02', '2024-01-02'], '제품': ['A', 'B', 'A', 'B'], '판매량': [100, 200, 150, 250], '이익': [20, 50, 30, 60]}
df = pd.DataFrame(data)
print(df.pivot(index='날짜', columns='제품', values=['판매량','이익']))
```
