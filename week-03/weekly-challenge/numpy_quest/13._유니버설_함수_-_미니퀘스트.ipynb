## 유니버설 함수 - 미니퀘스트


### 1번 퀘스트

배열 [1, 2, 3, 4]에 대해 NumPy 유니버설 함수를 사용하여
모든 요소를 제곱한 결과를 출력하는 코드를 작성하세요.


```python
import numpy as np
array = ([1,2,3,4])

print("np.power(array, 2):", np.power(array, 2))
```

### 2번 퀘스트

두 배열 [10, 20, 30], [1, 2, 3]의 요소별 합을 계산하고,
결과를 새로운 배열이 아닌 기존 배열에 저장하는 코드를 작성하세요.


```python
import numpy as np

array1 = np.array([10, 20, 30])
array2 = np.array([1, 2, 3])

array = np.add(array1, array2, out=array1)

print(array)
```

### 3번 퀘스트

배열 [1, e, 10, 100]에 대해 자연로그(log)를 계산하고,
log 값이 1보다 큰 요소만 출력하는 코드를 작성하세요.


```python
array = np.array([1, np.e, 10, 100])

big_log = np.log(array)

print(array[big_log > 1])

'''
값을 확인하는 법

import numpy as np
print(np.log(10))

output: 2.302...
'''
```
