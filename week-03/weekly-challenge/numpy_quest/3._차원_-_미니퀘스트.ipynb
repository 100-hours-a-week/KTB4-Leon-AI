### 1번 퀘스트

NumPy 배열 [[[1, 2], [3, 4]], [[5, 6], [7, 8]]]의
차원 수를 출력하는 코드를 작성하세요.


```python
import numpy as np
array = np.array([[[1, 2], [3, 4]], [[5, 6], [7, 8]]])

print(array.ndim)
```

### 2번 퀘스트

1차원 배열 [10, 20, 30, 40, 50, 60]을
(2, 3) 형태의 2차원 배열로 변환하여 출력하는 코드를 작성하세요.


```python
import numpy as np
array = np.array([10, 20, 30, 40, 50, 60])

print(array.reshape(2,3))
```

### 3번 퀘스트

1차원 배열 [7, 14, 21]에 newaxis를 사용하여 새로운 차원을 추가하고,
최종 배열의 모양과 차원 수를 출력하는 코드를 작성하세요.


```python
import numpy as np
array = np.array([7, 14, 21])
new_array = array[np.newaxis, :]

print(new_array.ndim)
```
