### 1번 퀘스트

NumPy 배열 [10, 20, 30]의 데이터 타입을 확인하는 코드를 작성하세요.


```python
import numpy as np

array= np.array([10,20,30])
print(array.dtype)
```

### 2번 퀘스트

정수형 배열 [1, 2, 3]을 부동소수점형 배열로 변환하고,
변환된 배열의 데이터 타입을 출력하는 코드를 작성하세요.


```python
import numpy as np

#int32
float_array = np.array([1,2,3], dtype=np.float64)
print(float_array.dtype)
```

### 3번 퀘스트

NumPy 배열 [100, 200, 300]을 uint8로 변환한 후,
메모리 사용량(바이트)을 출력하는 코드를 작성하세요.


```python
import numpy as np

array = np.array([100, 200, 300])
unit_array = array.astype(np.uint8)

print(unit_array.nbytes)
```
