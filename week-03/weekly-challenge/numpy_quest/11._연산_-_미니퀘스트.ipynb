### 1번 퀘스트

NumPy 배열 a = [1, 2, 3], b = [4, 5, 6]에 대해
요소별 덧셈을 수행하는 코드를 작성하세요.


```python
import numpy as np
a = np.array([1,2,3])
b = np.array([4,5,6])

print('a + b: ', a + b)
```
